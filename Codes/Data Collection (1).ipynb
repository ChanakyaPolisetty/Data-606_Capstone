```python
import yfinance
import warnings
warnings.filterwarnings("ignore") 
```

```python
engine = sqlalchemy.create_engine('mysql://root:12345678@localhost:3306/nasdaq')
for i in sc_df:
    df = yfinance.download(i,start='2017-01-01',end='2023-01-01')
    df = df.reset_index()
    df.to_sql(i,engine)
```
